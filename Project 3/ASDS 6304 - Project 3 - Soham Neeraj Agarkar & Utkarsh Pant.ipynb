## Q1.

Compute the gradient $\nabla f(x)$ and Hessian $\nabla^2f(x)$ of the Rosenbrock function $f(x) = 100(x_2 - x_{1}^2)^2 + (1 - x_1)^2$.

1. Show that $x^* = (1,1)^T$ is the only local minimizer of this function, and that the Hessian matrix at this point is positive definite.
2. Is the function convex or not? Justify your answer.
3. Plot the contour of this function for the interval chosen by yourself.

The given Rosenbrock function is $f(x) = 100(x_2 - x_1^2)^2 + (1 - x_1)^2$,

where $ x = \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} $. The gradient $ \nabla f(x) $ is a vector of partial derivatives:

$$
\nabla f(x) = \begin{bmatrix} \frac{\partial f}{\partial x_1} \\ \frac{\partial f}{\partial x_2} \end{bmatrix}
$$

**$\frac{\partial f}{\partial x_1}$**
   $= -400x_1(x_2 - x_1^2) - 2(1 - x_1)$
   $= -400x_1x_2 + 400x_1^3 - 2 + 2x_1$

**$\frac{\partial f}{\partial x_2}$**
   $= 200(x_2 - x_1^2)$

So, the gradient is:

$$
\nabla f(x) = \begin{bmatrix} -400x_1(x_2 - x_1^2) + 2x_1 - 2 \\ 200(x_2 - x_1^2) \end{bmatrix}
$$

**Calculating the Hessian**:

The Hessian $\nabla^2 f(x)$ is the matrix of second partial derivatives:

$$
\nabla^2 f(x) = \begin{bmatrix} \frac{\partial^2 f}{\partial x_1^2} & \frac{\partial^2 f}{\partial x_1 \partial x_2} \\ \frac{\partial^2 f}{\partial x_2 \partial x_1} & \frac{\partial^2 f}{\partial x_2^2} \end{bmatrix}
$$

**$\frac{\partial^2 f}{\partial x_1^2}$**:
   $= -400(x_2 - x_1^2) + 800x_1^2 + 2$


**$\frac{\partial^2 f}{\partial x_2^2}$**:
   $\frac{\partial^2 f}{\partial x_2^2} = 200$

**$\frac{\partial^2 f}{\partial x_1 \partial x_2}$**:
   $= -400x_1$

So, the Hessian is:

$$
\nabla^2 f(x) = \begin{bmatrix} -400(x_2 - x_1^2) + 800x_1^2 + 2 & -400x_1 \\ -400x_1 & 200 \end{bmatrix}
$$

### 1. Show that $x^* = (1, 1)^T$ is the only local minimizer of this function, and that the Hessian at this point is +ve semidefinite.

To show that $x^* = (1, 1)^T$ is the only local minimizer, we need to check the gradient and Hessian at this point.

#### Gradient at $x^* = (1, 1)^T$:

$$
\nabla f((1, 1)) = \begin{bmatrix} -400 \cdot 1 \cdot (1 - 1^2) + 2 \cdot 1 - 2 \\
200 \cdot (1 - 1^2) \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}
$$

The gradient is zero, indicating a critical point.

#### Hessian at $x^* = (1, 1)^T$:

$$
\nabla^2 f((1, 1)) = \begin{bmatrix} -400(1 - 1^2) + 800 \cdot 1^2 + 2 & -400 \cdot 1 \\
-400 \cdot 1 & 200 \end{bmatrix} = \begin{bmatrix} 802 & -400 \\ -400 & 200 \end{bmatrix}
$$

To check if the Hessian is positive definite, we need the eigenvalues of the matrix $\begin{bmatrix} 802 & -400 \\ -400 & 200 \end{bmatrix}$. The eigenvalues are positive, indicating a positive definite matrix.

Therefore, $x^* = (1, 1)^T$ is a local minimizer, and it's the only local minimizer for this function since the Rosenbrock function is known to have a unique global minimum at this point.

### 2. Convexity of the Function

The Rosenbrock function is **not convex**. Although it has a unique global minimum, the function is known to have a narrow, curved valley leading to the minimum. This characteristic makes the Hessian matrix not positive semi-definite everywhere in the domain, hence it is not globally convex.

### 3. Contour Plot of the Function

Here is the contour plot of the Rosenbrock function for the interval $x_1 \in [-2, 2]$ and $x_2 \in [-1, 3]$. The plot highlights the characteristic curved valley that leads to the global minimum at $(1, 1)$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Rosenbrock function
def rosenbrock(x, y):
    return 100 * (y - x**2)**2 + (1 - x)**2

# Generate a grid of points
x = np.linspace(-2, 2, 400)
y = np.linspace(-1, 3, 400)
X, Y = np.meshgrid(x, y)
Z = rosenbrock(X, Y)

# Plotting the contour
plt.style.use('default')
plt.figure(figsize=(10, 8))
contour = plt.contour(X, Y, Z, levels=np.logspace(-1, 3, 20), cmap='coolwarm')
plt.colorbar(contour)
plt.title('Contour Plot of the Rosenbrock Function')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.grid(True)
plt.show()
```

## Q2.

Show that the function $f(x) = 8x_1 + 12x_2 + x_{1}^2 − 2x_{2}^2$ has only one stationary point and that it is neither a maximum nor minimum, but a saddle point. Sketch the contour lines of $f$.

We can calculate the stationary point by calculating the gradient and setting it to __0__. To find the gradient, in turn, we need to calculate the partial derivative of $f$ w.r.t $x_1$ and $x_2$.


$$
\nabla f(x) = \begin{bmatrix} \frac{\partial f}{\partial x_1} \\
\frac{\partial f}{\partial x_2} \end{bmatrix}
$$

### Partial derivative with respect to $x_1$
   
$$
\frac{\partial f}{\partial x_1} = 8 + 2x_1
$$

### Partial derivative with respect to $x_2$
   
$$
\frac{\partial f}{\partial x_2} = 12 - 4x_2
$$

Therefore, we obtain the gradient matrix:

$$
\nabla f(x) = \begin{bmatrix} 8 + 2x_1 \\
                              12 - 4x_2 \end{bmatrix}
$$

### Calculating the stationary point:

Set the gradient to zero:

$$
\begin{cases}
8 + 2x_1 = 0 \\
12 - 4x_2 = 0
\end{cases}
$$

Solving these equations:

1. From $8 + 2x_1 = 0$, we get $x_1 = -4$.
2. From $12 - 4x_2 = 0$, we get $x_2 = 3$.

Thus, the stationary point is $(-4, 3)$. To determine whether this point is a local maximum, local minimum, or saddle point, we use the _Hessian matrix_, which we can calculate using second-order partial derivatives.

### Hessian Calculation

The Hessian matrix $\nabla^2 f(x)$ is:

$$
\nabla^2 f(x) = \begin{bmatrix} \frac{\partial^2 f}{\partial x_1^2} & \frac{\partial^2 f}{\partial x_1 \partial x_2} \\
\frac{\partial^2 f}{\partial x_2 \partial x_1} & \frac{\partial^2 f}{\partial x_2^2} \end{bmatrix}
$$

**Second derivative with respect to $x_1$**
   
$$
\frac{\partial^2 f}{\partial x_1^2} = 2
$$

**Second derivative with respect to $x_2$**
   
$$
\frac{\partial^2 f}{\partial x_2^2} = -4
$$

**Partial derivative w.r.t $x_1 x_2$**

$$
\frac{\partial^2 f}{\partial x_1 \partial x_2} = \frac{\partial^2 f}{\partial x_2 \partial x_1} = 0
$$

So, the Hessian matrix is:

$$
\nabla^2 f(x) = \begin{bmatrix} 2 & 0 \\ 0 & -4 \end{bmatrix}
$$

### Checking the determinant of the Hessian

The determinant of the Hessian is:

$$
\text{det}(\nabla^2 f(x)) = (2)(-4) - (0)(0) = -8
$$

The determiant being negative indicates that the critical point $(-4, 3)$ is a *saddle point.*

### Contour plot

```python
def f(x1, x2):
    return 8*x1 + 12*x2 + x1**2 - 2*x2**2

# mesh generated by evaluating the function at each point
x1 = np.linspace(-10, 10, 400)
x2 = np.linspace(-10, 10, 400)
X1, X2 = np.meshgrid(x1, x2)
Z = f(X1, X2)

# contour plot
plt.figure(figsize=(10, 8))
contour = plt.contour(X1, X2, Z, levels=np.linspace(-100, 100, 20), cmap='coolwarm')
plt.colorbar(contour)
plt.scatter(-4, 3, color='red', label='Saddle Point (-4, 3)', zorder=5, marker='x')
plt.title('Contour Plot of the Function $f(x) = 8x_1 + 12x_2 + x_1^2 - 2x_2^2$')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.grid(True)
plt.show()
```
